## NLP
It is a good practice to convert the strings to lowercase while performing analysis on them.

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
```

```python
print(en_stopwords)
```

```python
sentence = "It was too far to go to the shop and he did not want her to walk"
```

```python
lower_sentence = sentence.lower()
sentence_no_stop_words = ' '.join([word for word in lower_sentence.split(' ') if word not in en_stopwords])
```

### regular expression patterns

```python
customer_reviews = ["sam was happy with the product",
                   "amazing product - review by john",
                   "sarah says the toy was great",
                   "bike is very nice says matthew",
                   "cindy say the dress was beutiful"]
```

find review starting with letter a

```python
regex = "^a"
filtered_reviews = []
for review in customer_reviews:
    if re.search(regex, review):
        filtered_reviews.append(review)
print(filtered_reviews)
```

string ending with letter t

```python
regex = "t$"
filtered_reviews = []
for review in customer_reviews:
    if re.search(regex, review):
        filtered_reviews.append(review)
print(filtered_reviews)
```

Check multiple words in or

```python
regex = r"bike|dress"
filtered_reviews = []
for review in customer_reviews:
    if re.search(regex, review):
        filtered_reviews.append(review)
print(filtered_reviews)
```

Remove punctuations from the reviews

```python
regex = r"[^\w\s]"
filtered_reviews = []
for review in customer_reviews:
    filtered_reviews.append(re.sub(regex, "", review))
print(filtered_reviews)
```
