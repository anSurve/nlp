```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data = pd.read_csv("data/tripadvisor_hotel_reviews.csv")
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data['review_lowercase'][0]
```

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove('not')
```

```python
data['review_without_stopwords'] = data['review_lowercase'].apply(lambda a: ' '.join(word for word in a.split(' ') if word not in en_stopwords))
```

```python
data['review_without_stopwords'][0]
```

```python
punct_regex = r"[^\w\s\n]"
data['review_without_stopwords_and_puct'] = data.apply(lambda a: re.sub(r"[*]", 'star', a['review_without_stopwords']), axis=1)
data['review_without_stopwords_and_puct'] = data.apply(lambda a: re.sub(punct_regex, '', a['review_without_stopwords_and_puct']), axis=1)
```

```python
data.head()
```

```python
data['tokenize'] = data.apply(lambda a: word_tokenize(a['review_without_stopwords_and_puct']), axis=1)
```

```python
print(data['tokenize'][0])
```

```python
ps = PorterStemmer()
data['stemmed'] = data['tokenize'].apply(lambda a: [ps.stem(word) for word in a])
```

```python
print(data['stemmed'][0])
```

```python
lemmatizer = WordNetLemmatizer()
data['lemmatized'] = data['tokenize'].apply(lambda a: [lemmatizer.lemmatize(word) for word in a])
print(data['lemmatized'][0])
```

```python
# it combines all the lemmatized tokens from all the reviews combined
tokens_clean = sum(data['lemmatized'], [])
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
print(bigrams)
```

```python
trigrams = (pd.Series(nltk.ngrams(tokens_clean, 3)).value_counts())
print(trigrams)
```
