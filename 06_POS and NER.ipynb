```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bbc_data = pd.read_csv("data/bbc_news.csv")
bbc_data.head()
```

```python
titles = pd.DataFrame(bbc_data['title'])
```

```python
titles.head()
```

```python
titles['lowercase'] = titles['title'].str.lower()
titles.head()
```

```python
en_stopwords = stopwords.words('english')
titles['no_stopwords'] = titles['lowercase'].apply(lambda a: ' '.join(word for word in a.split(' ') if word not in en_stopwords))
titles.head()
```

```python
titles['no_punct'] = titles.apply(lambda x: re.sub(r"[^\w\s]", '', x['no_stopwords']), axis=1)
titles.head()
```

```python
titles['tokens_raw'] = titles.apply(lambda x: word_tokenize(x['title']), axis=1)
titles['tokens_clean'] = titles.apply(lambda x: word_tokenize(x['no_punct']), axis=1)
titles.head()
```

```python
lemmatizer = WordNetLemmatizer()
titles['tokens_clean_lemmatized'] = titles.apply(lambda x: [lemmatizer.lemmatize(word) for word in x['tokens_clean']], axis=1)
titles.head()
```

```python
tokens_raw_list = sum(titles['tokens_raw'], [])
tokens_clean_lemmatized_list = sum(titles['tokens_clean_lemmatized'], [])
```

### POS tagging

```python
nlp = spacy.load('en_core_web_sm')
```

```python
spacy_doc = nlp(' '.join(tokens_raw_list))
```

```python
pos_df = pd.DataFrame(columns=['token', 'pos_tag'])
```

```python
for token in spacy_doc:
    pos_df = pd.concat([
        pos_df, pd.DataFrame.from_records([{'token':token.text, 'pos_tag':token.pos_}])],ignore_index=True)
```

```python
pos_df_counts = pos_df.groupby(['token', 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

```python
pos_df_counts.head()
```

```python
# most common nouns
nouns = pos_df_counts[pos_df_counts.pos_tag == 'NOUN'][:10]
nouns
```

```python
# most common verbs
verbs = pos_df_counts[pos_df_counts.pos_tag == 'VERB'][:10]
verbs
```

```python
# most common adjectives
adjectives = pos_df_counts[pos_df_counts.pos_tag == 'ADJ'][:10]
adjectives
```

## NER 

```python
ner_df = pd.DataFrame(columns=['token', 'ner_tag'])
for token in spacy_doc.ents:
    if pd.isna(token.label_) is False:
        ner_df = pd.concat([ner_df,
                       pd.DataFrame.from_records([{'token': token.text, 'ner_tag':token.label_}])], ignore_index=True)
```

```python
ner_df.head()
```

```python
ner_df_counts = ner_df.groupby(['token', 'ner_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```
