```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

## TextBlob

```python
from textblob import TextBlob
```

```python
print(sentence_1)
sentiment_score_sentence_1 = TextBlob(sentence_1)
print(sentiment_score_sentence_1.polarity)
```

```python
print(sentence_2)
sentiment_score_sentence_2 = TextBlob(sentence_2)
print(sentiment_score_sentence_2.polarity)
```

```python
print(sentence_3)
sentiment_score_sentence_3 = TextBlob(sentence_3)
print(sentiment_score_sentence_3.polarity)
```

```python
print(sentence_4)
sentiment_score_sentence_4 = TextBlob(sentence_4)
print(sentiment_score_sentence_4.polarity)
```

## Vader

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
vaderSentiment = SentimentIntensityAnalyzer()
```

```python
print(sentence_1)
vader_sentiment_polarity_scores_sentence_1 = vaderSentiment.polarity_scores(sentence_1)
print(vader_sentiment_polarity_scores_sentence_1)
```

```python
print(sentence_2)
vader_sentiment_polarity_scores_sentence_2 = vaderSentiment.polarity_scores(sentence_2)
print(vader_sentiment_polarity_scores_sentence_2)
```

```python
print(sentence_3)
vader_sentiment_polarity_scores_sentence_3 = vaderSentiment.polarity_scores(sentence_3)
print(vader_sentiment_polarity_scores_sentence_3)
```

```python
print(sentence_4)
vader_sentiment_polarity_scores_sentence_4 = vaderSentiment.polarity_scores(sentence_4)
print(vader_sentiment_polarity_scores_sentence_4)
```

## Pre Trained Transformer Models 

```python
import transformers
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline('sentiment-analysis')
```

```python
print(sentence_1)
sentiment_pipeline(sentence_1)
```

```python
print(sentence_2)
sentiment_pipeline(sentence_2)
```

```python
print(sentence_3)
sentiment_pipeline(sentence_3)
```

```python
print(sentence_4)
sentiment_pipeline(sentence_4)
```
